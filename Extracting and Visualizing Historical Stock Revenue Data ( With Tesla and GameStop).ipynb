**#  Question  2 : Use Webscraping to Extract Tesla Revenue Data**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage containing Tesla's revenue data
url = "https://www.zippia.com/tesla-careers-11363/revenue/"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

tables = soup.find_all("table")

dataframe = []

for i, table in enumerate(tables):
    rows = table.find_all("tr")
    data = []
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]  # Iterating over each element to extract the text
        data.append(cols)
    df = pd.DataFrame(data)
    dataframe.append(df)    
    
dataframe[0].tail()  # Call head() with parentheses to display the first few rows
```

**#  Question 5 : Plot Tesla Stock Graph**

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'])
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Retrieve Tesla stock data
ticker = "TSLA"
tesla = yf.Ticker(ticker)
data = tesla.history(period="15d")

# Display the first five rows of data
print(data.head())

# Create the graph
title = "Tesla Stock Price (Last 15 Days)"
make_graph(data, title)
```
